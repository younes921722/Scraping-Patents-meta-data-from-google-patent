```python
import requests
import pandas as pd

def download_google_patents_csv(title, save_path):
    base_url = 'https://patents.google.com'
    search_url = f'{base_url}/xhr/query?url=q%3D({title})%26oq%3D{title}&exp=&download=true'

    # Send a GET request to Google Patents search page to get the CSV file
    response = requests.get(search_url)

    # Check if request was successful
    if response.status_code == 200:
        # Save the response content to a file
        with open(save_path, 'wb') as f:
            f.write(response.content)
        print(f"CSV file saved to {save_path}")
    else:
        print('Failed to retrieve CSV file')

# Example usage
title = "ai"
save_path = "patents.csv"
download_google_patents_csv(title, save_path)
```
